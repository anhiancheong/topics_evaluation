```python
"""
SUBSET THE DATA TO ONLY KEEP THE LAST 3 ROWS RIGHT NOW
THIS IS FOR EXPLORING ONLY THE PILOTED DATA
WE STILL WANT TO KEEP THE HEADERS THOUGH FOR REFERENCE
"""
observed_data = observed_data.drop([2,3,4,5])
intrusion_data = intrusion_data.drop([2,3,4,5,6,7])
 
```

```python
import numpy
def calculate_observed_precision(data):
    """
    Calculate the 'observed coherence' 'precision' score that is just the
    average value of the scores.
    data: Columns being explored
    """
    data_avg_total = []
    data_count = 0

    # Note: First 2 columns are 'header' information, all subsequent values are the actual values
    for d in data:
        values = [int(val) for val in data[d][2:] if not pd.isnull(val)]
        column_total = sum(values)
        column_avg = column_total / len(values)
        data_count += 1
        data_avg_total.append(column_avg)
        
    return sum(data_avg_total) / data_count, data_avg_total
        
    
print("News ", calculate_observed_precision(observed_data.iloc[:, 11:21]))
print("Leg ", calculate_observed_precision(observed_data.iloc[:, 21:41]))
print("Covid ", calculate_observed_precision(observed_data.iloc[:, 41:61]))
```

```python
def calculate_intrusion_precision(data):
    """
    0-1 - Number of 'correct' responses relative to the total
    """
    total_avg = []
    total_count = 0
    for question in data:
        values = [1 if int(val) == 6 else 0 for val in data[question][2:]]
        column_avg = sum(values) / len(values)

        total_avg.append(column_avg)
        total_count += 1
        
    return sum(total_avg) / total_count, total_avg
    
print("NEWS ", calculate_intrusion_precision(intrusion_data.iloc[:,11:21]))
print("LEG ", calculate_intrusion_precision(intrusion_data.iloc[:,21:41]))
print("COVID ", calculate_intrusion_precision(intrusion_data.iloc[:,41:61]))
```

```python
news_observed_total, news_observed_avgs = calculate_observed_precision(observed_data.iloc[:, 11:21])
leg_observed_total, leg_observed_avgs = calculate_observed_precision(observed_data.iloc[:, 21:41])
covid_observed_total, covid_observed_avgs = calculate_observed_precision(observed_data.iloc[:, 41:61])

news_intrusion_total, news_intrusion_avgs = calculate_intrusion_precision(intrusion_data.iloc[:,11:21])
leg_intrusion_total, leg_intrusion_avgs = calculate_intrusion_precision(intrusion_data.iloc[:,21:41])
covid_intrusion_total, covid_intrusion_avgs = calculate_intrusion_precision(intrusion_data.iloc[:,41:61])
```

```python
import scipy.stats
print(scipy.stats.pearsonr(news_observed_avgs, news_intrusion_avgs))
print(scipy.stats.pearsonr(leg_observed_avgs, leg_intrusion_avgs))
print(scipy.stats.pearsonr(covid_observed_avgs, covid_intrusion_avgs))
```
