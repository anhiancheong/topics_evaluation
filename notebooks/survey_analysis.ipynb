```python
import numpy
def calculate_observed_precision(data):
    """
    Calculate the 'observed coherence' 'precision' score that is just the
    average value of the scores.
    data: Columns being explored
    """
    data_avg_total = []
    data_count = 0

    # Note: First 2 columns are 'header' information, all subsequent values are the actual values
    for d in data:
        values = [int(val) for val in data[d][2:] if not pd.isnull(val)]
        column_total = sum(values)
        column_avg = column_total / len(values)
        data_count += 1
        data_avg_total.append(column_avg)
        
    return sum(data_avg_total) / data_count, data_avg_total
        
    
print("News ", calculate_observed_precision(observed_data.iloc[:, 11:21]))
print("Leg ", calculate_observed_precision(observed_data.iloc[:, 21:41]))
print("Covid ", calculate_observed_precision(observed_data.iloc[:, 41:61]))
```

```python
def calculate_intrusion_precision(data):
    """
    0-1 - Number of 'correct' responses relative to the total
    """
    total_avg = []
    total_count = 0
    for question in data:
        values = [1 if int(val) == 6 else 0 for val in data[question][2:]]
        column_avg = sum(values) / len(values)

        total_avg.append(column_avg)
        total_count += 1
        
    return sum(total_avg) / total_count, total_avg
    
print("NEWS ", calculate_intrusion_precision(intrusion_data.iloc[:,11:21]))
print("LEG ", calculate_intrusion_precision(intrusion_data.iloc[:,21:41]))
print("COVID ", calculate_intrusion_precision(intrusion_data.iloc[:,41:61]))
```

```python
from sklearn.metrics import cohen_kappa_score

def calculate_avg_cohen(df, col_start, col_end):
    total_pairs = 0
    total_kappa = 0.0

    for idx in range(0, len(df)):
        for idx_2 in range(idx + 1, len(df)):
            total_pairs += 1
            total_kappa += cohen_kappa_score(df.iloc[idx, col_start:col_end], df.iloc[idx_2, col_start:col_end])
            #print(cohen_kappa_score(df.iloc[idx, 11:61], df.iloc[idx_2, 11:61]))

    print("Total: ", total_kappa/total_pairs)
    return total_kappa/total_pairs
calculate_avg_cohen(observed_data, 11, 21)
calculate_avg_cohen(intrusion_data, 11, 21)
calculate_avg_cohen(observed_data, 21, 41)
calculate_avg_cohen(intrusion_data, 21, 41)
calculate_avg_cohen(observed_data, 41, 61)
calculate_avg_cohen(intrusion_data, 41, 61)
```
