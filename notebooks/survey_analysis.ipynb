```python
"""
Load in the list of topic model words
We'll 'merge' these with the collected data in order to merge better analyze future 
evaluations
"""
import csv
# Format of the coherence files is 'column' oriented, so we just need to reassmble things in the rows

def convert_topics_csv(file_handler, topic_model):
    lines = [line.replace("\n", "") for line in file_handler.readlines() if line]

    max_topic_idx = len(lines[0].split(","))
    # We only want the top 20 for covid and leg
    if topic_model in ["leg", "covid"]:
        max_topic_idx = 20
    
    topics = [{"topic_model": topic_model, "topic_id": "", "terms": [], "order_idx": None} for i in range(0, max_topic_idx)]
    line_count = 0
    for line in lines:
        components = line.split(",")[:max_topic_idx]

        
        #print(len(components), components)
        for idx, component in enumerate(components):
            if line_count == 0:
                topics[idx]["topic_id"] = component
                topics[idx]["order_idx"] = idx
            else:
                topics[idx]["terms"].append(component)
        line_count += 1
    return topics

news_topics = convert_topics_csv(open("/Users/hiancheong/Personal/grad/UMD/comp_ling/topics_evaluation/news_coherence.csv"), "news")
leg_topics = convert_topics_csv(open("/Users/hiancheong/Personal/grad/UMD/comp_ling/topics_evaluation/leg_coherence.csv"), "leg")
covid_topics = convert_topics_csv(open("/Users/hiancheong/Personal/grad/UMD/comp_ling/topics_evaluation/covid_coherence.csv"), "covid")
 
```

```python
import numpy
def calculate_observed_precision(data):
    """
    Calculate the 'observed coherence' 'precision' score that is just the
    average value of the scores.
    data: Columns being explored
    """
    data_avg_total = []
    data_count = 0

    # Note: First 2 columns are 'header' information, all subsequent values are the actual values
    for d in data:
        values = [int(val) for val in data[d][2:] if not pd.isnull(val)]
        column_total = sum(values)
        column_avg = column_total / len(values)
        data_count += 1
        data_avg_total.append(column_avg)
        
    return sum(data_avg_total) / data_count, data_avg_total
        
    
print("News ", calculate_observed_precision(observed_data.iloc[:, 11:21]))
print("Leg ", calculate_observed_precision(observed_data.iloc[:, 21:41]))
print("Covid ", calculate_observed_precision(observed_data.iloc[:, 41:61]))
```

```python
def calculate_intrusion_precision(data):
    """
    0-1 - Number of 'correct' responses relative to the total
    """
    total_avg = []
    total_count = 0
    for question in data:
        values = [1 if int(val) == 6 else 0 for val in data[question][2:]]
        column_avg = sum(values) / len(values)

        total_avg.append(column_avg)
        total_count += 1
        
    return sum(total_avg) / total_count, total_avg
    
print("NEWS ", calculate_intrusion_precision(intrusion_data.iloc[:,11:21]))
print("LEG ", calculate_intrusion_precision(intrusion_data.iloc[:,21:41]))
print("COVID ", calculate_intrusion_precision(intrusion_data.iloc[:,41:61]))
```

```python
from sklearn.metrics import cohen_kappa_score

def calculate_avg_cohen(df, col_start, col_end):
    total_pairs = 0
    total_kappa = 0.0

    for idx in range(0, len(df)):
        for idx_2 in range(idx + 1, len(df)):
            total_pairs += 1
            total_kappa += cohen_kappa_score(df.iloc[idx, col_start:col_end], df.iloc[idx_2, col_start:col_end])
            #print(cohen_kappa_score(df.iloc[idx, 11:61], df.iloc[idx_2, 11:61]))

    print("Total: ", total_kappa/total_pairs)
    return total_kappa/total_pairs
calculate_avg_cohen(observed_data, 11, 21)
calculate_avg_cohen(intrusion_data, 11, 21)
calculate_avg_cohen(observed_data, 21, 41)
calculate_avg_cohen(intrusion_data, 21, 41)
calculate_avg_cohen(observed_data, 41, 61)
calculate_avg_cohen(intrusion_data, 41, 61)
```

```python
from nltk import agreement

def calculate_krippensdorff(df):
    formatted_codes = []

    for question in df:
        for idx, val in enumerate(df[question]):
            formatted_codes.append((idx, question, val))

    ratingtask = agreement.AnnotationTask(data=formatted_codes)

    print('Krippendorff\'s alpha:',ratingtask.alpha())
    
    return ratingtask.alpha()

print("News Ratings")
calculate_krippensdorff(observed_data.iloc[:, 11:21])
print("Legislation Ratings")
calculate_krippensdorff(observed_data.iloc[:, 21:41])
print("Covid Ratings")
calculate_krippensdorff(observed_data.iloc[:, 41:61])

print("News Intrusion")
calculate_krippensdorff(intrusion_data.iloc[:, 11:21])
print("Legislation Intrusion")
calculate_krippensdorff(intrusion_data.iloc[:, 21:41])
print("Covid Intrusion")
calculate_krippensdorff(intrusion_data.iloc[:, 41:61])
```

```python
"""
Try and identify if there are any annotators that commonly disagree, this will help identify
any potentially bad data
"""
from collections import defaultdict

def find_agreement(data):
    data = [int(i) for i in data]
    annotator_count = defaultdict(int)
    for idx, value in enumerate(data):
        value =  int(value)
        sub_values = data[0:idx] + data[idx + 1:]
        mean = np.mean(sub_values)
        std = np.std(sub_values)
        
        #print(mean, std, value)
        
        if value >= mean + 2 * std or value <= mean - 2 * std:
            print("Outlier! "," Annotator ", idx, " Value was ", value, " Vs Mean: ", mean)
            annotator_count[idx] += 1
    return annotator_count
        
annotator_outliers = defaultdict(int)
for question in observed_data.iloc[:, 11:61]:
    outliers = find_agreement(observed_data[question])
    for key in outliers:
        annotator_outliers[key] += outliers[key]
```
